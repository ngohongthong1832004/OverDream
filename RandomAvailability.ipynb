```python
import pandas as pd
import random
import os

# Ensure directory exists for output
output_dir = "./availability/"
os.makedirs(output_dir, exist_ok=True)

# Read teacher data
teachers = pd.read_csv('./data/Data tự tạo - teacher.csv')

# Create timetable template
days = ["Thứ 2", "Thứ 3", "Thứ 4", "Thứ 5", "Thứ 6", "Thứ 7", "Chủ nhật"]
time_slots = ["Tiết 1-3", "Tiết 4-6", "Tiết 6-9", "Tiết 10-12", "Tiết 13-15"]

# Generate random availability
teacher_schedules = []
for _, teacher in teachers.iterrows():
    if teacher['Active'] == "0":
        # Teachers with Active == 0: Lower probability of being busy
        schedule = {
            day: [random.choices(["1", "0"], weights=[0.8, 0.2])[0] for _ in time_slots]
            for day in days
        }
    else:
        # Teachers with specific Active values: Equal probability
        schedule = {
            day: [random.choice(["1", "0"]) for _ in time_slots]
            for day in days
        }
    teacher_schedules.append({"Giáo viên": teacher['Tên '], **schedule})

# Convert to DataFrame and save to CSV
csv_file_paths = []
for schedule in teacher_schedules:
    df = pd.DataFrame(schedule)
    # df.insert(0, "Thời gian", time_slots)
    teacher_name = schedule["Giáo viên"].replace(" ", "_")
    file_path = os.path.join(output_dir, f"{teacher_name}_availability.csv")
    df.to_csv(file_path, index=False)
    csv_file_paths.append(file_path)

# Output generated file paths
csv_file_paths
```
