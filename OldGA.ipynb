```python
# Ensure 'Active' column is treated as a string
teacher['Active'] = teacher['Active'].astype(str)

# Split the 'Active' column into separate values and explode
teacher['Active'] = teacher['Active'].str.split(',')
teacher = teacher.explode('Active')

# Remove any leading/trailing whitespace in the 'Active' column
teacher['Active'] = teacher['Active'].str.strip()

# Handle cases where 'Active' is '0' (assign all courses)
teachers_with_all_courses = teacher[teacher['Active'] == '0']
teachers_with_specific_courses = teacher[teacher['Active'] != '0']

# Expand rows for teachers with 'Active' == '0' to include all courses
if not teachers_with_all_courses.empty:
    all_courses = subjects['Mã môn học'].tolist()
    expanded_rows = pd.DataFrame(
        [
            {'Tên ': row['Tên '], 'Active': course}
            for _, row in teachers_with_all_courses.iterrows()
            for course in all_courses
        ]
    )
    teachers_with_specific_courses = pd.concat([teachers_with_specific_courses, expanded_rows])

# Drop rows where 'Active' is empty or NaN
teachers_with_specific_courses = teachers_with_specific_courses.dropna(subset=['Active'])

# Remove empty strings in 'Active' column
teachers_with_specific_courses = teachers_with_specific_courses[teachers_with_specific_courses['Active'] != '']

# Ensure 'Active' is of the same type as 'Mã môn học'
teachers_with_specific_courses['Active'] = teachers_with_specific_courses['Active'].astype(subjects['Mã môn học'].dtype)

# Merge data with course codes
df_merged = pd.merge(teachers_with_specific_courses, subjects, left_on='Active', right_on='Mã môn học', how='inner')

# Display the merged DataFrame
df = df_merged[["Tên ", "Title", "Tên môn học", "Mã học phần"]]

print(df)

# Optionally, save the DataFrame to a file for further use
df_merged.to_csv('./data/merged_teacher_course_dataset.csv', index=False)
df.to_csv('./data/teacher_subject.csv', index=False)

```

```python
import pandas as pd
import random
import logging
import os

# Tạo logger riêng cho schedule generation
logger_schedule = logging.getLogger("ScheduleGeneration")
logger_schedule.setLevel(logging.INFO)

# Tạo handler riêng với mã hóa UTF-8
handler_schedule = logging.FileHandler('./logs/schedule_generation.log', mode='a', encoding='utf-8')
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler_schedule.setFormatter(formatter)

# Clear previous handlers to tránh log lặp
if logger_schedule.hasHandlers():
    logger_schedule.handlers.clear()

logger_schedule.addHandler(handler_schedule)

# Test ghi log
logger_schedule.info("Đã bắt đầu quá trình tạo thời khóa biểu.")

# Clear previous handlers to prevent duplicate logs
if logger_schedule.hasHandlers():
    logger_schedule.handlers.clear()

logger_schedule.addHandler(handler_schedule)

# Log thông báo
logger_schedule.info("Đã bắt đầu quá trình tạo thời khóa biểu.")

# Hàm khởi tạo thời khóa biểu
def GenSchedule(days, time_slots):
    try:
        # Bước 1: Đọc file subjects và lấy 10 môn học đầu tiên
        subjects_df = pd.read_csv('./data/Data tự tạo - subjects.csv')
        top_10_subjects = subjects_df.head(10)
        logger_schedule.info(f"Đọc 10 môn học đầu tiên thành công.")

        # Bước 2: Đọc file teacher_subject và lọc các giáo viên dạy môn đó
        teacher_subject_df = pd.read_csv('./data/teacher_subject.csv')
        filtered_teachers = teacher_subject_df[teacher_subject_df['Mã học phần'].isin(top_10_subjects['Mã học phần'])]
        logger_schedule.info(f"Lọc danh sách giáo viên dạy 10 môn học thành công.")

        # Bước 3: Tạo danh sách các cặp (giáo viên, môn học, tên môn học)
        subject_name_map = dict(zip(subjects_df['Mã học phần'], subjects_df['Tên môn học']))
        pairs = []
        for _, row in filtered_teachers.iterrows():
            teacher = row.get('Tên ', "Không xác định").strip()
            subject = row['Mã học phần']
            subject_name = subject_name_map.get(subject, "Môn học không xác định")
            pairs.append((teacher, subject, subject_name))
        logger_schedule.info(f"Tạo danh sách các cặp (giáo viên, môn học) thành công.")

        # Bước 4: Tạo thời khóa biểu trên tuần
        timetable = {day: {slot: None for slot in time_slots} for day in days}

        # Bước 5: Phân bố ngẫu nhiên các cặp vào timetable
        total_slots = len(days) * len(time_slots)
        pairs = pairs[:total_slots]  # Giới hạn số lượng cặp để phù hợp với số slot

        for pair in pairs:
            assigned = False
            for _ in range(100):  # Thử gán 100 lần trước khi bỏ qua cặp
                day = random.choice(days)
                slot = random.choice(time_slots)
                if timetable[day][slot] is None:
                    timetable[day][slot] = pair
                    assigned = True
                    break

            if not assigned:
                logger_schedule.warning(f"Không thể gán {pair} vào thời khóa biểu (không còn slot trống).")

        # Bước 6: Chuyển đổi thời khóa biểu sang dạng mảng
        timetable_array = []
        for day, slots in timetable.items():
            for slot, assigned in slots.items():
                if assigned:
                    teacher, subject, subject_name = assigned
                    timetable_array.append([day, slot, teacher, subject_name, subject])
                else:
                    timetable_array.append([day, slot, "Trống", "Trống", "Trống"])

        logger_schedule.info("Thời khóa biểu được tạo thành công.")
        return timetable_array

    except Exception as e:
        logger_schedule.error(f"Lỗi khi tạo thời khóa biểu: {e}")
        raise e  # Ném lỗi để xử lý ngoài hàm nếu cần

# Tạo quần thể thời khóa biểu
days = ["Thứ 2", "Thứ 3", "Thứ 4", "Thứ 5", "Thứ 6", "Thứ 7", "Chủ nhật"]
time_slots = ["Tiết 1-3", "Tiết 4-6", "Tiết 6-9", "Tiết 10-12", "Tiết 13-15"]

logger_schedule.info(f"Bắt đầu tạo quần thể với kích thước {1000}.")
POPULATION_SIZE = 1000
POPULATION = [GenSchedule(days, time_slots) for _ in range(POPULATION_SIZE)]

# Chuyển đổi quần thể thành DataFrame
flattened_population = []
for idx, schedule in enumerate(POPULATION):
    for entry in schedule:
        day, slot, teacher, subject_name, subject = entry
        flattened_population.append([idx + 1, day, slot, teacher, subject_name, subject])

# Tạo DataFrame từ quần thể
population_df = pd.DataFrame(flattened_population, columns=["Lịch ID", "Ngày", "Tiết", "Giáo viên", "Tên môn học", "Mã môn học"])

# Lưu vào file CSV
output_path = './data-GA/population.csv'
os.makedirs('./data-GA', exist_ok=True)  # Đảm bảo thư mục tồn tại
population_df.to_csv(output_path, index=False)
logger_schedule.info(f"Quần thể đã được lưu vào file: {output_path}")
```

### FITNESS FUNCTION

```python
import pandas as pd
import os
import logging

# Ensure logs directory exists
os.makedirs('./logs', exist_ok=True)

# Configure logging for fitness evaluation
logger_fitness = logging.getLogger("FitnessEvaluation")
logger_fitness.setLevel(logging.INFO)

# Clear previous handlers to avoid duplicate logs
if logger_fitness.hasHandlers():
    logger_fitness.handlers.clear()

# Create handler with UTF-8 encoding for logging to a file
handler_fitness = logging.FileHandler('./logs/fitness_evaluation.log', mode='a', encoding='utf-8')
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler_fitness.setFormatter(formatter)
logger_fitness.addHandler(handler_fitness)

# Log initialization
logger_fitness.info("Bắt đầu quá trình đánh giá fitness.")

# Function to standardize availability columns
def standardize_availability_columns(folder_path):
    for file in os.listdir(folder_path):
        if file.endswith(".csv"):
            try:
                file_path = os.path.join(folder_path, file)
                df = pd.read_csv(file_path)

                # Standardize column names
                df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
                df.columns = df.columns.str.replace("Chủ nhật", "Chủ nhật")  # Standardize names

                # Ensure all required columns are present
                required_columns = ["Giáo viên", "Thứ 2", "Thứ 3", "Thứ 4", "Thứ 5", "Thứ 6", "Thứ 7", "Chủ nhật"]
                missing_columns = [col for col in required_columns if col not in df.columns]
                for col in missing_columns:
                    df[col] = 0  # Add missing columns with default value 0

                # Reorder columns and save back to file
                df = df[required_columns]
                df.to_csv(file_path, index=False)
                logger_fitness.info(f"Đã chuẩn hóa file: {file}")
            except Exception as e:
                logger_fitness.error(f"Lỗi khi chuẩn hóa file {file}: {e}")

# Function to load availability data
def load_availability(folder_path):
    availability_data = {}
    for file in os.listdir(folder_path):
        if file.endswith(".csv"):
            try:
                teacher_name = file.replace("_availability.csv", "").replace("_", " ")
                file_path = os.path.join(folder_path, file)
                availability_data[teacher_name] = pd.read_csv(file_path)
                logger_fitness.info(f"Đã tải dữ liệu availability cho giáo viên: {teacher_name}")
            except Exception as e:
                logger_fitness.error(f"Lỗi khi tải dữ liệu availability từ file {file}: {e}")
    return availability_data

# Function to evaluate fitness for a single schedule
def evaluate_fitness(schedule, availability_data):
    fitness = 0
    for entry in schedule:
        day, slot, teacher, subject_name, subject = entry
        try:
            if teacher in availability_data:
                teacher_availability = availability_data[teacher]
                day_column = day

                if day_column not in teacher_availability.columns[1:]:
                    logger_fitness.warning(f"Cột {day_column} không tồn tại trong dữ liệu của giáo viên {teacher}")
                    continue

                slot_map = {"Tiết 1-3": 0, "Tiết 4-6": 1, "Tiết 6-9": 2, "Tiết 10-12": 3, "Tiết 13-15": 4}
                row_index = slot_map.get(slot, -1)

                if row_index == -1 or row_index >= len(teacher_availability):
                    logger_fitness.warning(f"Index {row_index} vượt quá số hàng trong dữ liệu của giáo viên {teacher}")
                    continue

                available = teacher_availability.iloc[row_index][day_column]
                if available == 1:
                    pass  # Increase fitness if teacher is available
                else:
                    fitness -= 1  # Decrease fitness if teacher is unavailable
            else:
                logger_fitness.warning(f"Không tìm thấy dữ liệu availability cho giáo viên {teacher}")
        except Exception as e:
            logger_fitness.error(f"Lỗi khi xử lý giáo viên {teacher}, ngày {day}, tiết {slot}: {e}")
    return fitness

# Function to evaluate fitness for the entire population
def evaluate_population(population, availability_folder):
    availability_data = load_availability(availability_folder)
    fitness_scores = []
    for idx, schedule in enumerate(population):
        try:
            fitness = evaluate_fitness(schedule, availability_data)
            fitness_scores.append({"Lịch ID": idx + 1, "Fitness": fitness})
            logger_fitness.info(f"Đã tính xong fitness cho Lịch ID: {idx + 1}")
        except Exception as e:
            logger_fitness.error(f"Lỗi khi tính fitness cho Lịch ID {idx + 1}: {e}")
    return pd.DataFrame(fitness_scores)

# Standardize availability files
availability_folder = "./availability"
standardize_availability_columns(availability_folder)

# Evaluate population fitness
fitness_scores = evaluate_population(POPULATION, availability_folder)

# Save fitness scores to a CSV file
output_file = "./data-GA/fitness_scores.csv"
try:
    os.makedirs('./data-GA', exist_ok=True)  # Ensure directory exists
    fitness_scores.to_csv(output_file, index=False)
    logger_fitness.info(f"Đã lưu kết quả fitness vào file: {output_file}")
except Exception as e:
    logger_fitness.error(f"Lỗi khi lưu kết quả fitness vào file: {e}")
```

```python
import pandas as pd
import random

# Load data
schedule_data = pd.read_csv('./data-GA/population.csv')  # Schedule data
fitness_scores = pd.read_csv('./data-GA/fitness_scores.csv')  # Fitness scores

# Helper function to load schedules by ID
def get_schedule_by_id(schedule_id, schedule_df, days, time_slots):
    schedule = {day: {slot: None for slot in time_slots} for day in days}
    rows = schedule_df[schedule_df["Lịch ID"] == schedule_id]
    for _, row in rows.iterrows():
        day, slot, teacher, subject_name, subject = row["Ngày"], row["Tiết"], row["Giáo viên"], row["Tên môn học"], row["Mã môn học"]
        schedule[day][slot] = (teacher, subject_name, subject)
    return schedule

# Selection: Roulette Wheel Selection
def select_parents(population, fitness_scores):
    total_fitness = sum(fitness_scores)
    probabilities = [1 - (score / total_fitness) for score in fitness_scores]
    selected_indices = random.choices(range(len(population)), weights=probabilities, k=2)
    return [population[i] for i in selected_indices]

# Crossover
def crossover(parent1, parent2, days, time_slots):
    crossover_point = random.randint(1, len(days) * len(time_slots) - 1)

    # Flatten parents
    parent1_flat = [entry for day in days for entry in parent1[day].values()]
    parent2_flat = [entry for day in days for entry in parent2[day].values()]

    # Perform crossover
    child1_flat = parent1_flat[:crossover_point] + parent2_flat[crossover_point:]
    child2_flat = parent2_flat[:crossover_point] + parent1_flat[crossover_point:]

    # Convert back to schedule format
    child1 = {day: {slot: child1_flat.pop(0) for slot in time_slots} for day in days}
    child2 = {day: {slot: child2_flat.pop(0) for slot in time_slots} for day in days}

    return child1, child2

# Mutation
def mutate(schedule, days, time_slots, mutation_rate=0.01):
    for day in days:
        for slot in time_slots:
            if random.random() < mutation_rate:
                teacher = random.choice(["Trống", "Nguyễn Văn A", "Trần Văn B"])  # Example teachers
                subject_name = random.choice(["Toán", "Lý", "Hóa"])  # Example subjects
                subject = "ID" + str(random.randint(1, 100))  # Example subject ID
                schedule[day][slot] = (teacher, subject_name, subject)
    return schedule

# Generate new population
def generate_new_population(population, fitness_scores, days, time_slots, crossover_rate=0.8, mutation_rate=0.01):
    new_population = []
    while len(new_population) < len(population):
        # Select parents
        parents = select_parents(population, fitness_scores)
        if random.random() < crossover_rate:
            # Perform crossover
            child1, child2 = crossover(parents[0], parents[1], days, time_slots)
        else:
            # Copy parents if no crossover
            child1, child2 = parents[0], parents[1]

        # Apply mutation
        # child1 = mutate(child1, days, time_slots, mutation_rate)
        # child2 = mutate(child2, days, time_slots, mutation_rate)
        
        

        # Add to new population
        new_population.append(child1)
        if len(new_population) < len(population):
            new_population.append(child2)

    return new_population

# Evaluate new fitness scores
def evaluate_population(population, evaluate_fitness_func):
    return [evaluate_fitness_func(individual) for individual in population]

# Evolution process
def evolve_population(schedule_data, fitness_scores, days, time_slots, generations=50):
    # Load initial population
    population = [get_schedule_by_id(i, schedule_data, days, time_slots) for i in range(1, len(fitness_scores) + 1)]
    fitness_list = fitness_scores["Fitness"].tolist()

    for generation in range(generations):
        print(f"Generation {generation + 1}")
        population = generate_new_population(population, fitness_list, days, time_slots)
        fitness_list = evaluate_population(population, evaluate_fitness_func)
        print(f"Best fitness in generation {generation + 1}: {max(fitness_list)}")

    # Return best schedule
    best_index = fitness_list.index(max(fitness_list))
    return population[best_index], max(fitness_list)

# Define days and time slots
days = ["Thứ 2", "Thứ 3", "Thứ 4", "Thứ 5", "Thứ 6", "Thứ 7", "Chủ nhật"]
time_slots = ["Tiết 1-3", "Tiết 4-6", "Tiết 6-9", "Tiết 10-12", "Tiết 13-15"]

# Define a dummy fitness evaluation function for demonstration

def evaluate_fitness_func(schedule):
    # Example: Fitness is the number of non-empty slots
    fitness = 0
    for day, slots in schedule.items():
        for entry in slots.values():
            if entry != None:
                fitness += 1
    return fitness

# Run evolution process
best_schedule, best_fitness = evolve_population(schedule_data, fitness_scores, days, time_slots)
print("Best Fitness:", best_fitness)
```
