```python
import pandas as pd
import random
import logging
import os

# Tạo logger riêng cho schedule generation
logger_schedule = logging.getLogger("ScheduleGeneration")
logger_schedule.setLevel(logging.INFO)

# Tạo handler riêng với mã hóa UTF-8
handler_schedule = logging.FileHandler('./logs/schedule_generation.log', mode='a', encoding='utf-8')
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler_schedule.setFormatter(formatter)

# Clear previous handlers to tránh log lặp
if logger_schedule.hasHandlers():
    logger_schedule.handlers.clear()

logger_schedule.addHandler(handler_schedule)

# Log thông báo
logger_schedule.info("Đã bắt đầu quá trình tạo thời khóa biểu.")

# Hàm khởi tạo thời khóa biểu
def GenSchedule(days, time_slots):
    try:
        # Bước 1: Đọc file subjects và lấy 10 môn học đầu tiên
        subjects_df = pd.read_csv('./data/Data tự tạo - subjects.csv')
        top_10_subjects = subjects_df.head(10)
        logger_schedule.info(f"Đọc 10 môn học đầu tiên thành công.")

        # Bước 2: Đọc file teacher_subject và lọc các giáo viên dạy môn đó
        teacher_subject_df = pd.read_csv('./data/teacher_subject.csv')
        filtered_teachers = teacher_subject_df[teacher_subject_df['Mã học phần'].isin(top_10_subjects['Mã học phần'])]
        logger_schedule.info(f"Lọc danh sách giáo viên dạy 10 môn học thành công.")

        # Bước 3: Tạo danh sách các cặp (giáo viên, môn học, tên môn học, số tín chỉ)
        subject_info_map = dict(zip(subjects_df['Mã học phần'], zip(subjects_df['Tên môn học'], subjects_df['Số tín chỉ'])))
        pairs = []
        for _, row in filtered_teachers.iterrows():
            teacher = row.get('Tên ', "Không xác định").strip()
            subject = row['Mã học phần']
            subject_name, credits = subject_info_map.get(subject, ("Môn học không xác định", 0))
            pairs.append((teacher, subject, subject_name, credits))
        logger_schedule.info(f"Tạo danh sách các cặp (giáo viên, môn học, tín chỉ) thành công.")

        # Bước 4: Tạo thời khóa biểu trên tuần
        timetable = {day: {slot: None for slot in time_slots} for day in days}

        # Bước 5: Phân bố ngẫu nhiên các cặp vào timetable
        total_slots = len(days) * len(time_slots)
        pairs = pairs[:total_slots]  # Giới hạn số lượng cặp để phù hợp với số slot

        for pair in pairs:
            assigned = False
            for _ in range(100):  # Thử gán 100 lần trước khi bỏ qua cặp
                day = random.choice(days)
                slot = random.choice(time_slots)
                if timetable[day][slot] is None:
                    timetable[day][slot] = pair
                    assigned = True
                    break

            if not assigned:
                logger_schedule.warning(f"Không thể gán {pair} vào thời khóa biểu (không còn slot trống).")

        # Bước 6: Chuyển đổi thời khóa biểu sang dạng mảng
        timetable_array = []
        for day, slots in timetable.items():
            for slot, assigned in slots.items():
                if assigned:
                    teacher, subject, subject_name, credits = assigned
                    timetable_array.append([day, slot, teacher, subject_name, subject, credits])
                else:
                    timetable_array.append([day, slot, "Trống", "Trống", "Trống", 0])

        logger_schedule.info("Thời khóa biểu được tạo thành công.")
        return timetable_array

    except Exception as e:
        logger_schedule.error(f"Lỗi khi tạo thời khóa biểu: {e}")
        raise e  # Ném lỗi để xử lý ngoài hàm nếu cần

# Tạo quần thể thời khóa biểu
days = ["Thứ 2", "Thứ 3", "Thứ 4", "Thứ 5", "Thứ 6", "Thứ 7", "Chủ nhật"]
time_slots = ["Tiết 1-3", "Tiết 4-6", "Tiết 6-9", "Tiết 10-12", "Tiết 13-15"]

logger_schedule.info(f"Bắt đầu tạo quần thể với kích thước {1000}.")
POPULATION_SIZE = 1000
POPULATION = [GenSchedule(days, time_slots) for _ in range(POPULATION_SIZE)]

# Chuyển đổi quần thể thành DataFrame
flattened_population = []
for idx, schedule in enumerate(POPULATION):
    for entry in schedule:
        day, slot, teacher, subject_name, subject, credits = entry
        flattened_population.append([idx + 1, day, slot, teacher, subject_name, subject, credits])

# Tạo DataFrame từ quần thể
population_df = pd.DataFrame(flattened_population, columns=["Lịch ID", "Ngày", "Tiết", "Giáo viên", "Tên môn học", "Mã môn học", "Số tín chỉ"])

# Lưu vào file CSV
output_path = './data-GA/population.csv'
os.makedirs('./data-GA', exist_ok=True)  # Đảm bảo thư mục tồn tại
population_df.to_csv(output_path, index=False)
logger_schedule.info(f"Quần thể đã được lưu vào file: {output_path}")
```

## FILTER

```python
import pandas as pd

# Đọc dữ liệu từ file
schedule_data = pd.read_csv('./data-GA/population.csv')  # Đường dẫn tới file gốc

def filter_schedule(data):
    """
    Lọc số tiết sao cho mỗi môn học trong từng lịch ID thỏa mãn:
    - Nếu số tín chỉ > 2: Tối đa 2 tiết trên tuần.
    - Nếu số tín chỉ <= 2: Tối đa 1 tiết trên tuần.
    """
    # Đọc dữ liệu từ DataFrame
    df = pd.DataFrame(data)
    
    # Sử dụng groupby để nhóm theo "Lịch ID" và "Mã môn học"
    def filter_group(group):
        max_sessions = 2 if group['Số tín chỉ'].iloc[0] > 2 else 1
        return group.sample(min(len(group), max_sessions))
    
    df_filtered = df.groupby(["Lịch ID", "Mã môn học"], group_keys=False).apply(filter_group)
    
    return df_filtered

# Lọc dữ liệu
filtered_schedule = filter_schedule(schedule_data)

# Ghi kết quả vào file mới
filtered_schedule.to_csv('./data-GA/filtered_schedule.csv', index=False, encoding='utf-8-sig')

# Hiển thị thông báo hoàn thành
print("Lịch học đã được lọc theo số tín chỉ và lưu vào file './data-GA/filtered_schedule.csv'")
```

### TRỌNG SỐ GIẢNG VIÊN ÍT VÀO TRƯỜNG ( Càng nhỏ càng tốt ) 

```python
import pandas as pd

def calculate_average_weight(data):
    """
    Tính trọng số trung bình của từng nhóm lịch ID dựa trên quy luật giảng viên đi dạy.
    
    Quy tắc:
    - Mỗi buổi dạy của giảng viên cộng thêm 3 điểm trọng số.
    - Tính trung bình trọng số của tất cả các giảng viên trong một lịch ID.
    """
    # Đọc dữ liệu từ DataFrame
    df = pd.DataFrame(data)
    
    # Đếm số buổi dạy của mỗi giảng viên theo từng lịch ID
    teacher_count = df.groupby(['Lịch ID', 'Giáo viên']).size().reset_index(name='Số buổi')
    
    # Tính trọng số cho mỗi giảng viên: mỗi buổi dạy cộng 3 điểm
    teacher_count['Trọng số'] = teacher_count['Số buổi'] * 3
    
    # Tính tổng trọng số và số giảng viên cho từng Lịch ID
    weight_summary = teacher_count.groupby('Lịch ID').agg(
        Tổng_trọng_số=('Trọng số', 'sum'),
        Số_giáo_viên=('Giáo viên', 'count')
    ).reset_index()
    
    # Tính trọng số trung bình
    weight_summary['Trọng số trung bình'] = (weight_summary['Tổng_trọng_số'] / weight_summary['Số_giáo_viên'])
    
    return weight_summary[['Lịch ID', 'Trọng số trung bình']]

# Đọc dữ liệu từ file CSV (hoặc sử dụng dữ liệu trực tiếp)
schedule_data = pd.read_csv('./data-GA/filtered_schedule.csv')  # Đường dẫn tới file gốc

# Tính trọng số trung bình
average_weights = calculate_average_weight(schedule_data)

# Hiển thị kết quả
print(average_weights)

# Ghi kết quả vào file nếu cần
average_weights.to_csv('./data-GA/average_weights.csv', index=False, encoding='utf-8-sig')
```

### TRỌNG SỐ LỊCH RẢNH CỦA GIẢNG VIÊN ( Càng nhỏ càng tốt )

```python
import pandas as pd
import numpy as np
import random
import os
import gc
import matplotlib.pyplot as plt

# Định nghĩa lớp cá thể
class Individual:
    def __init__(self, schedule):
        self.schedule = schedule  # Lịch học của cá thể (DataFrame)
        self.fitness = None  # Giá trị fitness của cá thể

# Hàm fitness thứ nhất
def calculate_average_weight(schedule):
    df = schedule.copy()
    df.columns = df.columns.str.strip()

    required_columns = ['Lịch ID', 'Giáo viên']
    for col in required_columns:
        if col not in df.columns:
            raise KeyError(f"Cột '{col}' không tồn tại trong DataFrame.")

    teacher_count = df.groupby(['Lịch ID', 'Giáo viên']).size().reset_index(name='Số buổi')
    teacher_count['Trọng số'] = teacher_count['Số buổi'] * 3
    weight_summary = teacher_count.groupby('Lịch ID').agg(
        Tổng_trọng_số=('Trọng số', 'sum'),
        Số_giáo_viên=('Giáo viên', 'count')
    ).reset_index()
    weight_summary['Trọng số trung bình'] = weight_summary['Tổng_trọng_số'] / weight_summary['Số_giáo_viên']
    return weight_summary['Trọng số trung bình'].mean()

# Hàm load dữ liệu rảnh của giảng viên
def load_availability(folder_path):
    availability_data = {}

    try:
        for file_name in os.listdir(folder_path):
            if file_name.endswith(".csv"):
                file_path = os.path.join(folder_path, file_name)
                df = pd.read_csv(file_path)
                df.columns = df.columns.str.strip()

                if df.empty or 'Giáo viên' not in df.columns:
                    print(f"File {file_name} không đúng định dạng hoặc không có dữ liệu.")
                    continue

                for teacher, group in df.groupby('Giáo viên'):
                    if teacher not in availability_data:
                        availability_data[teacher] = group.reset_index(drop=True)
                    else:
                        availability_data[teacher] = pd.concat(
                            [availability_data[teacher], group.reset_index(drop=True)],
                            ignore_index=True
                        )
    except Exception as e:
        print(f"Lỗi khi load dữ liệu availability: {e}")

    return availability_data

# Hàm fitness thứ hai
def evaluate_availability_fitness(schedule, availability_data):
    schedule = schedule.copy()
    schedule.columns = schedule.columns.str.strip()

    required_columns = ['Ngày', 'Tiết', 'Giáo viên']
    for col in required_columns:
        if col not in schedule.columns:
            raise KeyError(f"Cột '{col}' không tồn tại trong DataFrame.")

    teacher_weights = {}
    slot_map = {
        "Tiết 1-3": 0,
        "Tiết 4-6": 1,
        "Tiết 7-9": 2,
        "Tiết 10-12": 3,
        "Tiết 13-15": 4,
    }

    for idx, row in schedule.iterrows():
        day = row['Ngày'].strip()
        slot = row['Tiết'].replace(' ', '')
        teacher = row['Giáo viên']

        if teacher not in availability_data or day not in availability_data[teacher].columns:
            continue

        row_index = slot_map.get(slot, -1)
        if row_index == -1 or row_index >= len(availability_data[teacher]):
            continue

        if teacher not in teacher_weights:
            teacher_weights[teacher] = {"weight": 0, "count": 0}

        availability = availability_data[teacher].iloc[row_index, availability_data[teacher].columns.get_loc(day)]
        teacher_weights[teacher]["weight"] += 3 if availability == 0 else 0
        teacher_weights[teacher]["count"] += 1

    total_weight = sum(data["weight"] for data in teacher_weights.values())
    total_count = sum(data["count"] for data in teacher_weights.values())
    return total_weight / total_count if total_count > 0 else 0

# Hàm tính fitness tổng hợp
def calculate_total_fitness(individual, availability_data):
    fitness1 = calculate_average_weight(individual.schedule)
    fitness2 = evaluate_availability_fitness(individual.schedule, availability_data)
    return fitness1 * 0.3 + fitness2 * 0.7

# Hàm filter_schedule (from your code)
def filter_schedule(data):
    """
    Lọc số tiết sao cho mỗi môn học trong từng lịch ID thỏa mãn:
    - Nếu số tín chỉ > 2: Tối đa 2 tiết trên tuần.
    - Nếu số tín chỉ <= 2: Tối đa 1 tiết trên tuần.
    """
    df = pd.DataFrame(data)
    
    def filter_group(group):
        max_sessions = 2 if group['Số tín chỉ'].iloc[0] > 2 else 1
        if len(group) > max_sessions:
            return group.sample(n=max_sessions)
        else:
            return group
    
    df_filtered = df.groupby(["Lịch ID", "Mã môn học"], group_keys=False).apply(filter_group)
    
    return df_filtered

# Hàm đột biến
def mutate_schedule(schedule):
    schedule = schedule.copy()
    num_rows = len(schedule)
    if num_rows > 1:
        idx1, idx2 = np.random.choice(num_rows, size=2, replace=False)
        schedule.at[idx1, 'Giáo viên'], schedule.at[idx2, 'Giáo viên'] = (
            schedule.at[idx2, 'Giáo viên'],
            schedule.at[idx1, 'Giáo viên'],
        )
    # Apply filter_schedule to ensure constraints are maintained
    schedule = filter_schedule(schedule)
    return schedule

# Hàm crossover
def uniform_crossover_schedule(parent1_schedule, parent2_schedule):
    child_schedule = parent1_schedule.copy()
    for i in range(len(child_schedule)):
        if random.random() < 0.5:
            child_schedule.at[i, 'Giáo viên'] = parent2_schedule.at[i, 'Giáo viên']
    # Apply filter_schedule to ensure constraints are maintained
    child_schedule = filter_schedule(child_schedule)
    return child_schedule

# Hàm chọn lọc
def tournament_selection(population, k):
    selected = random.sample(population, k)
    selected.sort(key=lambda ind: ind.fitness)
    return selected[0]

# Hàm tiến hóa quần thể
def evolve_population(population, generations, mutation_rate, crossover_rate, availability_data):
    best_fitness_over_time = []
    for generation in range(generations):
        # Đánh giá fitness
        for individual in population:
            individual.fitness = calculate_total_fitness(individual, availability_data)

        # Sắp xếp theo fitness
        population.sort(key=lambda ind: ind.fitness)
        best_individual = population[0]
        best_fitness_over_time.append(best_individual.fitness)
        print(f"Generation {generation + 1}, Best Fitness: {best_individual.fitness}")

        # Tạo quần thể mới
        new_population = [best_individual]  # Elitism

        while len(new_population) < len(population):
            # Chọn lọc
            parent1 = tournament_selection(population, k=3)
            parent2 = tournament_selection(population, k=3)

            # Crossover
            if random.random() < crossover_rate:
                child_schedule = uniform_crossover_schedule(parent1.schedule, parent2.schedule)
            else:
                child_schedule = parent1.schedule.copy()

            # Đột biến
            if random.random() < mutation_rate:
                child_schedule = mutate_schedule(child_schedule)

            # Thêm cá thể mới
            new_population.append(Individual(child_schedule))

        population = new_population

    return best_individual, best_fitness_over_time

# Hàm vẽ lịch
def plot_schedule(schedule, filename):
    days = ['Thứ 2', 'Thứ 3', 'Thứ 4', 'Thứ 5', 'Thứ 6', 'Thứ 7', 'Chủ nhật']
    time_slots = ['Tiết 1-3', 'Tiết 4-6', 'Tiết 7-9', 'Tiết 10-12', 'Tiết 13-15']
    schedule_matrix = pd.DataFrame('', index=time_slots, columns=days)

    for idx, row in schedule.iterrows():
        day = row['Ngày'].strip()
        slot = row['Tiết'].strip()
        teacher = row['Giáo viên'].strip()
        subject = row['Tên môn học'].strip()
        if day in days and slot in time_slots:
            schedule_matrix.at[slot, day] = f"{subject}\n{teacher}"

    # Vẽ bảng thời khóa biểu
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.axis('off')
    table = ax.table(cellText=schedule_matrix.values,
                     rowLabels=schedule_matrix.index,
                     colLabels=schedule_matrix.columns,
                     cellLoc='center',
                     loc='upper left')
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 1.2)
    plt.savefig(filename)
    plt.close()

# Tham số thuật toán
population_size = 20
generations = 50
mutation_rate = 0.1
crossover_rate = 0.7

# Đọc dữ liệu lịch học ban đầu
initial_schedule = pd.read_csv('./data-GA/filtered_schedule.csv')
initial_schedule.columns = initial_schedule.columns.str.strip()

# Áp dụng bộ lọc vào lịch học ban đầu
initial_schedule = filter_schedule(initial_schedule)
initial_schedule.to_csv('./data-GA/initial_schedule.csv', index=False, encoding='utf-8-sig')

# Tạo quần thể ban đầu
population = [Individual(mutate_schedule(initial_schedule)) for _ in range(population_size)]
availability_data = load_availability('./availability')

# Tiến hóa quần thể
best_individual, best_fitness_over_time = evolve_population(
    population, generations, mutation_rate, crossover_rate, availability_data
)

# Lưu kết quả
print("Best individual fitness:", best_individual.fitness)
best_individual.schedule.to_csv('./data-GA/best_schedule.csv', index=False, encoding='utf-8-sig')

# Lưu dữ liệu để vẽ hàm hội tụ
best_fitness_df = pd.DataFrame({
    'Generation': range(1, generations + 1),
    'Best Fitness': best_fitness_over_time
})
best_fitness_df.to_csv('./data-GA/best_fitness_over_time.csv', index=False)

# Vẽ hàm hội tụ
plt.figure()
plt.plot(best_fitness_df['Generation'], best_fitness_df['Best Fitness'], marker='o')
plt.title('Convergence over Generations')
plt.xlabel('Generation')
plt.ylabel('Best Fitness')
plt.savefig('./data-GA/convergence_plot.png')
plt.close()

# Vẽ thời khóa biểu tốt nhất
plot_schedule(best_individual.schedule, './data-GA/best_schedule_plot.png')
```
