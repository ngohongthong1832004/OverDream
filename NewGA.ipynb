```python
# Ensure 'Active' column is treated as a string
teacher['Active'] = teacher['Active'].astype(str)

# Split the 'Active' column into separate values and explode
teacher['Active'] = teacher['Active'].str.split(',')
teacher = teacher.explode('Active')

# Remove any leading/trailing whitespace in the 'Active' column
teacher['Active'] = teacher['Active'].str.strip()

# Handle cases where 'Active' is '0' (assign all courses)
teachers_with_all_courses = teacher[teacher['Active'] == '0']
teachers_with_specific_courses = teacher[teacher['Active'] != '0']

# Expand rows for teachers with 'Active' == '0' to include all courses
if not teachers_with_all_courses.empty:
    all_courses = subjects['Mã môn học'].tolist()
    expanded_rows = pd.DataFrame(
        [
            {'Tên ': row['Tên '], 'Active': course}
            for _, row in teachers_with_all_courses.iterrows()
            for course in all_courses
        ]
    )
    teachers_with_specific_courses = pd.concat([teachers_with_specific_courses, expanded_rows])

# Drop rows where 'Active' is empty or NaN
teachers_with_specific_courses = teachers_with_specific_courses.dropna(subset=['Active'])

# Remove empty strings in 'Active' column
teachers_with_specific_courses = teachers_with_specific_courses[teachers_with_specific_courses['Active'] != '']

# Ensure 'Active' is of the same type as 'Mã môn học'
teachers_with_specific_courses['Active'] = teachers_with_specific_courses['Active'].astype(subjects['Mã môn học'].dtype)

# Merge data with course codes
df_merged = pd.merge(teachers_with_specific_courses, subjects, left_on='Active', right_on='Mã môn học', how='inner')

# Display the merged DataFrame
df = df_merged[["Tên ", "Title", "Tên môn học", "Mã học phần"]]

print(df)

# Optionally, save the DataFrame to a file for further use
df_merged.to_csv('./data/merged_teacher_course_dataset.csv', index=False)
df.to_csv('./data/teacher_subject.csv', index=False)

```

```python
import pandas as pd
import random
import logging
import os

# Tạo logger riêng cho schedule generation
logger_schedule = logging.getLogger("ScheduleGeneration")
logger_schedule.setLevel(logging.INFO)

# Tạo handler riêng với mã hóa UTF-8
handler_schedule = logging.FileHandler('./logs/schedule_generation.log', mode='a', encoding='utf-8')
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler_schedule.setFormatter(formatter)

# Clear previous handlers to tránh log lặp
if logger_schedule.hasHandlers():
    logger_schedule.handlers.clear()

logger_schedule.addHandler(handler_schedule)

# Log thông báo
logger_schedule.info("Đã bắt đầu quá trình tạo thời khóa biểu.")

# Hàm khởi tạo thời khóa biểu
def GenSchedule(days, time_slots):
    try:
        # Bước 1: Đọc file subjects và lấy 10 môn học đầu tiên
        subjects_df = pd.read_csv('./data/Data tự tạo - subjects.csv')
        top_10_subjects = subjects_df.head(10)
        logger_schedule.info(f"Đọc 10 môn học đầu tiên thành công.")

        # Bước 2: Đọc file teacher_subject và lọc các giáo viên dạy môn đó
        teacher_subject_df = pd.read_csv('./data/teacher_subject.csv')
        filtered_teachers = teacher_subject_df[teacher_subject_df['Mã học phần'].isin(top_10_subjects['Mã học phần'])]
        logger_schedule.info(f"Lọc danh sách giáo viên dạy 10 môn học thành công.")

        # Bước 3: Tạo danh sách các cặp (giáo viên, môn học, tên môn học, số tín chỉ)
        subject_info_map = dict(zip(subjects_df['Mã học phần'], zip(subjects_df['Tên môn học'], subjects_df['Số tín chỉ'])))
        pairs = []
        for _, row in filtered_teachers.iterrows():
            teacher = row.get('Tên ', "Không xác định").strip()
            subject = row['Mã học phần']
            subject_name, credits = subject_info_map.get(subject, ("Môn học không xác định", 0))
            pairs.append((teacher, subject, subject_name, credits))
        logger_schedule.info(f"Tạo danh sách các cặp (giáo viên, môn học, tín chỉ) thành công.")

        # Bước 4: Tạo thời khóa biểu trên tuần
        timetable = {day: {slot: None for slot in time_slots} for day in days}

        # Bước 5: Phân bố ngẫu nhiên các cặp vào timetable
        total_slots = len(days) * len(time_slots)
        pairs = pairs[:total_slots]  # Giới hạn số lượng cặp để phù hợp với số slot

        for pair in pairs:
            assigned = False
            for _ in range(100):  # Thử gán 100 lần trước khi bỏ qua cặp
                day = random.choice(days)
                slot = random.choice(time_slots)
                if timetable[day][slot] is None:
                    timetable[day][slot] = pair
                    assigned = True
                    break

            if not assigned:
                logger_schedule.warning(f"Không thể gán {pair} vào thời khóa biểu (không còn slot trống).")

        # Bước 6: Chuyển đổi thời khóa biểu sang dạng mảng
        timetable_array = []
        for day, slots in timetable.items():
            for slot, assigned in slots.items():
                if assigned:
                    teacher, subject, subject_name, credits = assigned
                    timetable_array.append([day, slot, teacher, subject_name, subject, credits])
                else:
                    timetable_array.append([day, slot, "Trống", "Trống", "Trống", 0])

        logger_schedule.info("Thời khóa biểu được tạo thành công.")
        return timetable_array

    except Exception as e:
        logger_schedule.error(f"Lỗi khi tạo thời khóa biểu: {e}")
        raise e  # Ném lỗi để xử lý ngoài hàm nếu cần

# Tạo quần thể thời khóa biểu
days = ["Thứ 2", "Thứ 3", "Thứ 4", "Thứ 5", "Thứ 6", "Thứ 7", "Chủ nhật"]
time_slots = ["Tiết 1-3", "Tiết 4-6", "Tiết 6-9", "Tiết 10-12", "Tiết 13-15"]

logger_schedule.info(f"Bắt đầu tạo quần thể với kích thước {1000}.")
POPULATION_SIZE = 5000
POPULATION = [GenSchedule(days, time_slots) for _ in range(POPULATION_SIZE)]

# Chuyển đổi quần thể thành DataFrame
flattened_population = []
for idx, schedule in enumerate(POPULATION):
    for entry in schedule:
        day, slot, teacher, subject_name, subject, credits = entry
        flattened_population.append([idx + 1, day, slot, teacher, subject_name, subject, credits])

# Tạo DataFrame từ quần thể
population_df = pd.DataFrame(flattened_population, columns=["Lịch ID", "Ngày", "Tiết", "Giáo viên", "Tên môn học", "Mã môn học", "Số tín chỉ"])

# Lưu vào file CSV
output_path = './data-GA/population.csv'
os.makedirs('./data-GA', exist_ok=True)  # Đảm bảo thư mục tồn tại
population_df.to_csv(output_path, index=False)
logger_schedule.info(f"Quần thể đã được lưu vào file: {output_path}")
```

## FILTER

```python
import pandas as pd

# Đọc dữ liệu từ file
schedule_data = pd.read_csv('./data-GA/population.csv')  # Đường dẫn tới file gốc

def filter_schedule(data):
    """
    Lọc số tiết sao cho mỗi môn học trong từng lịch ID thỏa mãn:
    - Nếu số tín chỉ > 2: Tối đa 2 tiết trên tuần.
    - Nếu số tín chỉ <= 2: Tối đa 1 tiết trên tuần.
    """
    # Đọc dữ liệu từ DataFrame
    df = pd.DataFrame(data)
    
    # Sử dụng groupby để nhóm theo "Lịch ID" và "Mã môn học"
    def filter_group(group):
        max_sessions = 2 if group['Số tín chỉ'].iloc[0] > 2 else 1
        return group.sample(min(len(group), max_sessions))
    
    df_filtered = df.groupby(["Lịch ID", "Mã môn học"], group_keys=False).apply(filter_group)
    
    return df_filtered

# Lọc dữ liệu
filtered_schedule = filter_schedule(schedule_data)

# Ghi kết quả vào file mới
filtered_schedule.to_csv('./data-GA/filtered_schedule.csv', index=False, encoding='utf-8-sig')

# Hiển thị thông báo hoàn thành
print("Lịch học đã được lọc theo số tín chỉ và lưu vào file './data-GA/filtered_schedule.csv'")
```

### TRỌNG SỐ GIẢNG VIÊN ÍT VÀO TRƯỜNG

```python
import pandas as pd

def calculate_weight(data):
    """
    Tính trọng số của từng nhóm lịch ID dựa trên quy luật giảng viên đi dạy.
    
    Quy tắc:
    - Nếu giảng viên dạy 1 buổi trong tuần, cộng thêm 3 điểm trọng số.
    - Tính trung bình trọng số của tất cả các giảng viên trên một lịch ID.
    """
    # Đọc dữ liệu từ DataFrame
    df = pd.DataFrame(data)
    
    # Đếm số buổi dạy của mỗi giảng viên theo từng lịch ID
    teacher_count = df.groupby(['Lịch ID', 'Giáo viên']).size().reset_index(name='Số buổi')
    
    # Tính trọng số cho mỗi giảng viên
    teacher_count['Trọng số'] = teacher_count['Số buổi'].apply(lambda x: 3 if x == 1 else 0)
    
    # Tính trọng số trung bình cho từng Lịch ID
    average_weights = teacher_count.groupby('Lịch ID')['Trọng số'].mean().reset_index(name='Trọng số trung bình')
    
    return average_weights

# Đọc dữ liệu từ file CSV (hoặc sử dụng dữ liệu trực tiếp)
schedule_data = pd.read_csv('./data-GA/filtered_schedule.csv')  # Đường dẫn tới file gốc

# Tính trọng số trung bình
average_weights = calculate_weight(schedule_data)

# Hiển thị kết quả
print(average_weights)

# Ghi kết quả vào file nếu cần
average_weights.to_csv('./data-GA/average_weights.csv', index=False, encoding='utf-8-sig')
```

```python
# Function to evaluate fitness for a single schedule
def evaluate_fitness(schedule, availability_data):
    """
    Tính fitness dựa trên lịch rảnh của từng giảng viên và trả về trung bình fitness của lịch.
    """
    teacher_fitness = {}
    
    # Tính fitness cho từng giảng viên trong lịch
    for entry in schedule:
        if len(entry) >= 3:  # Đảm bảo định dạng đúng
            day, slot, teacher = entry[:3]
        else:
            logger_fitness.error(f"Số lượng cột không đúng: {entry}")
            continue

        if teacher not in teacher_fitness:
            teacher_fitness[teacher] = {"fitness": 0, "days_in_school": set()}

        try:
            if teacher in availability_data:
                teacher_availability = availability_data[teacher]
                day_column = day

                if day_column not in teacher_availability.columns[1:]:
                    logger_fitness.warning(f"Cột {day_column} không tồn tại trong dữ liệu của giáo viên {teacher}")
                    continue

                # Mapping slot to row index
                slot_map = {"Tiết 1-3": 0, "Tiết 4-6": 1, "Tiết 6-9": 2, "Tiết 10-12": 3, "Tiết 13-15": 4}
                row_index = slot_map.get(slot, -1)

                if row_index == -1 or row_index >= len(teacher_availability):
                    logger_fitness.warning(f"Index {row_index} vượt quá số hàng trong dữ liệu của giáo viên {teacher}")
                    continue

                available = teacher_availability.iloc[row_index][day_column]
                if available == 1:
                    teacher_fitness[teacher]["fitness"] += 1  # Tăng fitness nếu giáo viên rảnh
                else:
                    teacher_fitness[teacher]["fitness"] -= 1  # Giảm fitness nếu giáo viên không rảnh

                # Ghi nhận ngày vào trường
                teacher_fitness[teacher]["days_in_school"].add(day)
            else:
                logger_fitness.warning(f"Không tìm thấy dữ liệu availability cho giáo viên {teacher}")
        except Exception as e:
            logger_fitness.error(f"Lỗi khi xử lý giáo viên {teacher}, ngày {day}, tiết {slot}: {e}")

    # Tính trung bình fitness của lịch dựa trên tất cả các giảng viên
    total_fitness = sum([data["fitness"] for data in teacher_fitness.values()])
    return total_fitness / len(teacher_fitness) if teacher_fitness else 0

# Function to evaluate fitness for the entire population
def evaluate_population(population, availability_folder):
    """
    Tính fitness cho toàn bộ quần thể dựa trên fitness trung bình của từng lịch.
    """
    availability_data = load_availability(availability_folder)
    fitness_scores = []
    for idx, schedule in enumerate(population):
        try:
            fitness = evaluate_fitness(schedule, availability_data)
            fitness_scores.append({"Lịch ID": idx + 1, "Fitness": fitness})
            logger_fitness.info(f"Đã tính xong fitness cho Lịch ID: {idx + 1}")
        except Exception as e:
            logger_fitness.error(f"Lỗi khi tính fitness cho Lịch ID {idx + 1}: {e}")
    return pd.DataFrame(fitness_scores)

# Standardize availability files
availability_folder = "./availability"
standardize_availability_columns(availability_folder)

# Evaluate population fitness
fitness_scores = evaluate_population(POPULATION, availability_folder)

# Save fitness scores to a CSV file
output_file = "./data-GA/fitness_scores.csv"
try:
    os.makedirs('./data-GA', exist_ok=True)  # Ensure directory exists
    fitness_scores.to_csv(output_file, index=False)
    logger_fitness.info(f"Đã lưu kết quả fitness vào file: {output_file}")
except Exception as e:
    logger_fitness.error(f"Lỗi khi lưu kết quả fitness vào file: {e}")
```

```python
import pandas as pd
import random

def combine_fitness_weights(fitness_df, weight_df):
    """
    Gộp dữ liệu Fitness và Trọng số trung bình.
    """
    combined_df = pd.merge(fitness_df, weight_df, on='Lịch ID')
    return combined_df

def mutate_schedule(schedule_data, combined_df, num_mutations=10):
    """
    Thực hiện đột biến lịch:
    - Kết hợp lịch có Fitness cao với Fitness thấp.
    - Randomize lịch giữ nguyên ràng buộc.
    """
    # Sắp xếp lịch theo Fitness và Trọng số trung bình
    high_fitness = combined_df.sort_values(['Fitness', 'Trọng số trung bình'], ascending=[False, False])
    low_fitness = combined_df.sort_values(['Fitness', 'Trọng số trung bình'], ascending=[True, True])

    # Lấy danh sách Lịch ID từ cả hai nhóm
    high_ids = high_fitness['Lịch ID'].tolist()
    low_ids = low_fitness['Lịch ID'].tolist()

    # Chọn ngẫu nhiên các cặp lịch ID để đột biến
    mutations = []
    for _ in range(num_mutations):
        high_id = random.choice(high_ids)
        low_id = random.choice(low_ids)

        # Lấy dữ liệu của từng lịch ID
        high_schedule = schedule_data[schedule_data['Lịch ID'] == high_id]
        low_schedule = schedule_data[schedule_data['Lịch ID'] == low_id]

        # Đột biến: Kết hợp random các tiết học từ cả hai lịch
        mutated_schedule = pd.concat([
            high_schedule.sample(len(high_schedule) // 2, random_state=random.randint(1, 100)),
            low_schedule.sample(len(low_schedule) // 2, random_state=random.randint(1, 100))
        ])

        # Đảm bảo giữ nguyên ràng buộc số tiết, môn học
        mutated_schedule = mutated_schedule.drop_duplicates(subset=['Tiết', 'Mã môn học']).reset_index(drop=True)
        mutated_schedule['Lịch ID'] = max(schedule_data['Lịch ID']) + len(mutations) + 1  # Tạo ID mới
        mutations.append(mutated_schedule)

    # Kết hợp các lịch đột biến vào dữ liệu ban đầu
    mutated_data = pd.concat([schedule_data] + mutations, ignore_index=True)
    return mutated_data

# Dữ liệu Fitness và Trọng số trung bình
fitness_data = pd.read_csv('./data-GA/fitness_scores.csv')  # Đọc file gốc

weight_data = pd.read_csv('./data-GA/average_weights.csv')  # Đọc file gốc

# Dữ liệu lịch học
schedule_data = pd.read_csv('./data-GA/filtered_schedule.csv')  # Đọc file gốc

# Gộp dữ liệu Fitness và Trọng số trung bình
combined_df = combine_fitness_weights(fitness_data, weight_data)

# Đột biến lịch
mutated_data = mutate_schedule(schedule_data, combined_df, num_mutations=5)

# Lưu kết quả ra file
mutated_data.to_csv('./data-GA/mutated_schedule.csv', index=False, encoding='utf-8-sig')
print("Lịch đột biến đã được lưu vào './data-GA/mutated_schedule.csv'")
```

# TIẾN HÓA QUẦN THỂ

```python
import pandas as pd
import random
import logging

# Tạo logger riêng cho schedule generation
logger_schedule = logging.getLogger("ScheduleGeneration")
logger_schedule.setLevel(logging.INFO)

# Tạo handler riêng với mã hóa UTF-8
handler_schedule = logging.FileHandler('./logs/evolution_generation.log', mode='a', encoding='utf-8')
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler_schedule.setFormatter(formatter)

# Clear previous handlers để tránh log lặp
if logger_schedule.hasHandlers():
    logger_schedule.handlers.clear()

logger_schedule.addHandler(handler_schedule)

logger_schedule.info("Đã bắt đầu quá trình tạo thời khóa biểu.")


def combine_fitness_weights(fitness_df, weight_df):
    """
    Gộp dữ liệu Fitness và Trọng số trung bình.
    """
    logger_schedule.info("Checking columns in fitness_df and weight_df...")
    logger_schedule.debug(f"Fitness DataFrame columns: {fitness_df.columns}")
    logger_schedule.debug(f"Weight DataFrame columns: {weight_df.columns}")

    if 'Lịch ID' not in fitness_df.columns or 'Lịch ID' not in weight_df.columns:
        logger_schedule.error("Cột 'Lịch ID' không tồn tại trong một hoặc cả hai DataFrame.")
        raise KeyError("Cột 'Lịch ID' không tồn tại trong một hoặc cả hai DataFrame.")

    fitness_df['Lịch ID'] = fitness_df['Lịch ID'].astype(int)
    weight_df['Lịch ID'] = weight_df['Lịch ID'].astype(int)

    combined_df = pd.merge(fitness_df, weight_df, on='Lịch ID', how='inner')
    return combined_df


def mutate_schedule(schedule_data, combined_df, num_mutations=10):
    """
    Thực hiện đột biến lịch.
    """
    high_fitness = combined_df.sort_values(['Fitness', 'Trọng số trung bình'], ascending=[False, False])
    low_fitness = combined_df.sort_values(['Fitness', 'Trọng số trung bình'], ascending=[True, True])

    high_ids = high_fitness['Lịch ID'].tolist()
    low_ids = low_fitness['Lịch ID'].tolist()

    mutations = []
    for _ in range(num_mutations):
        if not high_ids or not low_ids:
            logger_schedule.warning("Không đủ lịch để đột biến.")
            break

        high_id = random.choice(high_ids)
        low_id = random.choice(low_ids)

        high_schedule = schedule_data[schedule_data['Lịch ID'] == high_id]
        low_schedule = schedule_data[schedule_data['Lịch ID'] == low_id]

        if high_schedule.empty or low_schedule.empty:
            logger_schedule.warning("Lịch không hợp lệ. Bỏ qua đột biến.")
            continue

        mutated_schedule = pd.concat([
            high_schedule.sample(len(high_schedule) // 2, random_state=random.randint(1, 100)),
            low_schedule.sample(len(low_schedule) // 2, random_state=random.randint(1, 100))
        ])

        mutated_schedule = mutated_schedule.drop_duplicates(subset=['Tiết', 'Mã môn học']).reset_index(drop=True)
        mutated_schedule['Lịch ID'] = max(schedule_data['Lịch ID']) + len(mutations) + 1
        mutations.append(mutated_schedule)

    mutated_data = pd.concat([schedule_data] + mutations, ignore_index=True)
    return mutated_data


def evolve_schedule(schedule_data, fitness_data, weight_data, num_generations=500, num_mutations=10):
    """
    Hàm tiến hóa để tối ưu hóa lịch giảng dạy.
    """
    current_schedule = schedule_data.copy()

    for generation in range(num_generations):
        logger_schedule.info(f"Thế hệ {generation + 1}/{num_generations}")

        # Kết hợp Fitness và Trọng số trung bình
        combined_df = combine_fitness_weights(fitness_data, weight_data)

        if combined_df.empty:
            logger_schedule.warning("Combined DataFrame is empty. Skipping this generation.")
            continue

        # Đột biến lịch
        mutated_data = mutate_schedule(current_schedule, combined_df, num_mutations)

        # Đánh giá lại Fitness và Trọng số
        fitness_data = evaluate_population(mutated_data, './availability')
        if fitness_data.empty:
            logger_schedule.warning(f"Fitness DataFrame is empty at generation {generation + 1}. Using default values.")
            fitness_data = pd.DataFrame({'Lịch ID': mutated_data['Lịch ID'].unique(), 'Fitness': [0] * len(mutated_data['Lịch ID'].unique())})

        weight_data = calculate_weight(mutated_data)
        if weight_data.empty:
            logger_schedule.warning(f"Weight DataFrame is empty at generation {generation + 1}. Using default values.")
            weight_data = pd.DataFrame({'Lịch ID': mutated_data['Lịch ID'].unique(), 'Trọng số trung bình': [0] * len(mutated_data['Lịch ID'].unique())})

        combined_df = combine_fitness_weights(fitness_data, weight_data)

        # Lọc lịch với trọng số cao hơn trung bình
        threshold = combined_df['Trọng số trung bình'].mean()
        logger_schedule.debug(f"Threshold for filtering: {threshold}")

        filtered_schedule_ids = combined_df[combined_df['Trọng số trung bình'] > threshold]['Lịch ID']
        if filtered_schedule_ids.empty:
            logger_schedule.warning(f"No schedules above threshold {threshold}. Skipping this generation.")
            continue

        current_schedule = mutated_data[mutated_data['Lịch ID'].isin(filtered_schedule_ids)]

        if combined_df.empty:
            logger_schedule.warning("No valid schedules available for best selection.")
            continue

        # Chọn lịch tốt nhất
        best_schedule = combined_df.sort_values(['Fitness', 'Trọng số trung bình'], ascending=[False, False]).iloc[0]
        logger_schedule.info(f"Lịch tốt nhất sau thế hệ {generation + 1}: {best_schedule}")

    best_combined = combine_fitness_weights(fitness_data, weight_data)
    if best_combined.empty:
        logger_schedule.error("No valid schedules available after all generations.")
        return None

    best_schedule_id = best_combined.sort_values(['Fitness', 'Trọng số trung bình'], ascending=[False, False]).iloc[0]['Lịch ID']
    final_schedule = current_schedule[current_schedule['Lịch ID'] == best_schedule_id]
    logger_schedule.info("Lịch tiến hóa cuối cùng đã được tạo thành công.")
    return final_schedule



# Đọc dữ liệu từ file CSV
schedule_data = pd.read_csv('./data-GA/filtered_schedule.csv')
fitness_data = pd.read_csv('./data-GA/fitness_scores.csv')
weight_data = pd.read_csv('./data-GA/average_weights.csv')

logger_schedule.info("Bắt đầu tiến hóa thời khóa biểu...")
final_schedule = evolve_schedule(schedule_data, fitness_data, weight_data, num_generations=500, num_mutations=10)
```
